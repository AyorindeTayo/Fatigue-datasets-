```python
import pandas as pd
import numpy as np

# Load your CSV dataset
data = pd.read_csv('Fatigue data.csv')

# Define the number of augmentations to create for each data point
num_augmentations = 500  # You can adjust this number as needed

# Create an empty list to store augmented data
augmented_data = []

# Perform data augmentation
for _, row in data.iterrows():
    for _ in range(num_augmentations):
        augmented_row = row.copy()

        # Add random noise to numeric columns with only positive values
        for col in augmented_row.index:
            if pd.api.types.is_numeric_dtype(augmented_row[col]):
                augmented_row[col] = augmented_row[col] + np.abs(np.random.normal(0, 0.1))  # Adjust the standard deviation as needed

        # Add slight perturbations to categorical columns
        for col in augmented_row.index:
            if pd.api.types.is_categorical_dtype(augmented_row[col]):
                unique_values = augmented_row[col]
                if len(unique_values) > 1:
                    perturbation = np.random.choice(unique_values, 1)[0]
                    augmented_row[col] = perturbation

        augmented_data.append(augmented_row)

# Combine the original data and augmented data
augmented_data = pd.DataFrame(augmented_data)
augmented_dataset = pd.concat([data, augmented_data], ignore_index=True)

# Save the augmented dataset to a new CSV file
augmented_dataset.to_csv('Augmented_Fatigue_data.csv', index=False)
```
